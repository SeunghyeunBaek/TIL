```python
import bs4
import requests
import time
```

```python
url = 'https://pjt3591oo.github.io/'
# 페이지 요청해서 응답결과를 받는다.
res = requests.get(url)
print(res) # 200
#print(res.text)
# bs4 객체 만들기
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
# div , class = p
div_list = soup.find_all('div',class_='p')
# 셀렉터1 셀렉터2 : 셀럭터 1에 해당하는 태그 내부의 태그 중 셀렉터 2에 해당하는 태그 선택
# 셀렉터1>셀렉터2 : 셀럭터1에 해당하는 태그 내부의 하위(자식) 태그 중 셀렉터 2에 해당하는 태그 선택
# 셀렉터1셀럭터2 : 셀렉터1과 셀렉터2를 동시에 만족하는 태그 선택

div_list_s = soup.select('div.p')
for tag in div_list:
    print(tag.prettify())
    print('-'*100)
```

```python
data_list = []

# 태그의 수만큼 반복한다.
for tag in div_list:
    h3 = tag.find('h3')
    # print(h3)
    a = h3.find('a')
    # print(a)
    href = a.attrs['href']
    # print(href)
    # 문자열 데이터
    a_text = a.text
    # print(a_text)
    # h4 태그를 가져온다
    h4 = tag.find('h4')
    h4_text = h4.text.strip()
    #print(h4_text)
    p_tag = tag.find('p',class_='author')
    # print(p_tag)
    p_tag_span = p_tag.find('span',class_='date')
    #print(p_tag_span)
    p_tag_span_text = p_tag_span.text.strip()
    span_tag_list = p_tag_span_text.split('|')
    date_text = span_tag_list[0]
    author_text = span_tag_list[1]
    #print(f'{date_text},{author_text}')

    #추출 데이터를 DIC 에 답는다.
    data_dic = {
        'href':href,
        'a_text':a_text,
        'h4_text':h4_text,
        'date_text':date_text
    }
    #리스트에 담는다.
    data_list.append(data_dic)
```

```python
for data in data_list:
    print(f'href : {data.get("href")}')
    print(f'a_text : {data.get("a_text")}')
    print(f'h4_text : {data.get("h4_text")}')
    print(f'date_text : {data.get("date_text")}')
    print('='*100)
```

```python
# 주소를 받아 요청하고, div tage p class list를 반환한다.
def getDIvList(site):
    res = requests.get(site)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    divList = soup.find_all('div',class_='p')
    
    # 다음 페이지가 있는지 확인한다.
    # Next 라는 문자열 값을 가지고 있는 태그를 가져온다.
    next_tag = soup.find(any,text='Next') # tag를 any 로 지정해서 모든 태그를 검색한다.
    next_target = None
    if next_tag.name == 'a':
        next_target = next_tag.attrs['href'].jujmjm`` 
    return divList,next_target
```

```python
# div_list 를 받아서 data_list 를 반환한다.
def getData(tag):
    href = tag.select('h3>a')[0].attrs['href'].strip()
    a = tag.select('h3>a')[0].text.strip()
    h4 = tag.select('h4')[0].text.strip()
    tmp = tag.select('p.author>span.date')[0].text.strip()
    tmp = tmp.split('|')
    date = tmp[0]
    author = tmp[1] 
    return href,a,h4,author
```

```python
# 리스트에 담긴 데이터를 출력
def printList(data_list):
    for data in data_list:
        for key,value in data.items():
            print(f'{key}:{value}')
        print('='*100)
```
