# 5. 스칼라 선택 더 빠르게 하기
`.iat`, `.at` 으로 더 빠르게 scalar 선택 가능


```python
college_df = pd.read_csv(root + file_name, index_col='INSTNM')
```

```python
# loc
college_name_selected = 'Texas A & M University-College Station'
attr_name_selected = 'UGDS_WHITE'

college_df.loc[college_name_selected, attr_name_selected]
```

```python
# at
college_df.at[college_name_selected, attr_name_selected]
```

```python
# 수행시간 비교
%timeit college_df.loc[college_name_selected, attr_name_selected]
```

```python
%timeit college_df.at[college_name_selected, attr_name_selected]
```

```python
# iloc vs iat
row_num = college_df.index.get_loc(college_name_selected)
col_num = college_df.columns.get_loc(attr_name_selected)
```

```python
%timeit college_df.iloc[row_num, col_num]
```

```python
%timeit college_df.iat[row_num, col_num]
```
