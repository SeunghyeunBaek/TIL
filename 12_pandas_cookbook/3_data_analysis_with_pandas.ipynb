## 데이터 형식을 변경해 저장
* categorical 변수 중 숫자로 표현됐고 범위가 적을 경우 형변환을 통해(int64 -> int8) 메모리 절약

```python
col_names = ['RELAFFIL', 'SATMTMID', 'CURROPER', 'INSTNM', 'STABBR']
cols_df = college_df.loc[:,col_names]
cols_df.head()
```

```python
# check memory usage and data type
original_type = cols_df.dtypes
orignal_mem = cols_df.memory_usage().copy()
pd.DataFrame([original_type, orignal_mem]).T.rename(columns = {1: 'dtype', 0: 'memory_usage'})
```

```python
# memory 감소
cols_df['RELAFFIL'] = cols_df['RELAFFIL'] .astype(np.int8)  # int64 -> int8
cols_df['STABBR'] = cols_df['STABBR'].astype('category')  # unique value의 수가 적은 열을 object -> category 변환

trans_type = cols_df.dtypes
trans_mem = cols_df.memory_usage().copy()
pd.DataFrame([trans_type, trans_mem]).T.rename(columns = {1: 'dtype', 0: 'memory_usage'})
```

## 최대에서 최소 선택
* nlargest: 상위 n개
* msmallest: 하위 n개

```python
file_name = 'movie.csv'
movie_df = pd.read_csv(root + file_name)
col_selected = ['movie_title', 'imdb_score', 'budget']
movie2_df = movie_df.loc[:, col_selected]
movie2_df.head()
```

```python
# imdbscore 상위 100개 영화 추출
movie2_df.nlargest(n = 100, columns = 'imdb_score').head()
```

```python
# 상위 imdb_score 상위 100개 영화 중 최저예산 영화 5개
movie2_df.nlargest(n=100, columns='imdb_score').nsmallest(5, 'budget')
```

## 정렬을 통해 각 그룹에서 최대 선택


```python
# 각 해 최고 평점
col_selected = ['movie_title', 'title_year', 'imdb_score']
movie3_df = movie_df.loc[:, col_selected]
movie3_df = movie3_df.sort_values(by=['title_year', 'imdb_score'], ascending=False)  # 연도, 점수기준 내림차순 정렬
movie_top_year = movie3_df.drop_duplicates(subset='title_year', keep='first')  # title_year 가 같은 row 중 첫번째 row만 남기고 삭제 
movie_top_year.head()
```

```python
# 각 해 최저 평점
movie4_df = movie3_df.sort_values(by=['title_year', 'imdb_score'], ascending=[False, True])  # ascending list로 입력
movie_worst_year = movie4_df.drop_duplicates(subset='title_year', keep='first')
movie_worst_year.head()
```

## funtions

```python
def init_eda(df):
    '''print initial informations for input dataframe
    Args:
        df (pd.DataFrame): input dataFrame
    returns:
        1. head of dataframe
        2. shape of dataframe
        3. dtype, number of not null values per column
        4. descriptive statistics
            4.1.: numeric feature: count, min, std, min, qutile, max
            4.2.: categorical features: number of unique values, most frequent value, top frequency'''
    
    print('===== 1. head =====')
    display(df.head())
    
    print('===== 2. shape =====')
    print(df.shape)
    
    print('===== 3. dtype, number of not null values, memory usage =====')
    display(df.info())
    
    print('===== 4. decriptive statistics =====')
    
    print('===== 4.1. numeric features =====')
    desc_df = df.describe().T 
    
    null_ratio = df.isnull().mean()  # null value ratio
    null_ratio_idx = desc_df.columns.get_loc('count') + 1
    
    desc_df.insert(loc=null_ratio_idx,
                    column='null_ratio',
                    value=null_ratio)  # insert null value ratio next to count

    display(desc_df)
    
    print('===== 4.2. categorical features =====')
    display(df.describe(include=[np.object, pd.Categorical]).T)
    
    return None
```
