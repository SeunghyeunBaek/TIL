```python
def init_eda(df):
    '''print initial informations for input dataframe
    Args:
        df (pd.DataFrame): input dataFrame
    returns:
        1. head of dataframe
        2. shape of dataframe
        3. dtype, number of not null values per column
        4. descriptive statistics
            4.1.: numeric feature: count, min, std, min, qutile, max
            4.2.: categorical features: number of unique values, most frequent value, top frequency'''
    
    print('===== 1. head =====')
    display(df.head())
    
    print('===== 2. shape =====')
    print(df.shape)
    
    print('===== 3. dtype, number of not null values, memory usage =====')
    display(df.info())
    
    print('===== 4. decriptive statistics =====')
    
    print('===== 4.1. numeric features =====')
    desc_df = df.describe().T 
    
    null_ratio = df.isnull().mean()  # null value ratio
    null_ratio_idx = desc_df.columns.get_loc('count') + 1
    
    desc_df.insert(loc=null_ratio_idx,
                    column='null_ratio',
                    value=null_ratio)  # insert null value ratio next to count

    display(desc_df)
    
    print('===== 4.2. categorical features =====')
    display(df.describe(include=[np.object, pd.Categorical]).T)
    
    return None
```
