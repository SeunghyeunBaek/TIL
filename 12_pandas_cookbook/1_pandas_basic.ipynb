## rename index

```python
# 리스트로 컬럼, 열 값 지정하기
# df.shape과 리스트의 길이가 동일할 때 가능

movie_df = pd.read_csv(root + file_name, index_col='movie_title')
idx = movie_df.index
columns = movie_df.columns

idx_list = idx.tolist()
column_list = columns.tolist()

idx_list[0] = 'changed_idx'
column_list[0] = 'changed_column'

movie_df.index = idx_list
movie_df.columns = column_list

movie_df.head(3)
```

## 열 생성,삭제

```python
movie_df['has_seen'] = 0
movie_df['has_seen'].head(3)
```

```python
movie_df['sum_facebook_likes'] = \
movie_df['actor_1_facebook_likes'] + \
movie_df['actor_2_facebook_likes'] + \
movie_df['actor_3_facebook_likes'] + \
movie_df['director_facebook_likes']
```

```python
# 연산하는 Series 중 누락값이 있을 경우 0으로 처리함
movie_df['sum_facebook_likes'].fillna(0, inplace=True)
```

```python
# cast_total_facebook_lkkes: 모든 출연진, 제작자의 좋아요 합
movie_df['director_actor_likes_ratio'] = \
movie_df['sum_facebook_likes'] / movie_df['cast_total_facebook_likes']

movie_df['director_actor_likes_ratio'].head(20)
```

```python
movie_df.columns.get_loc('gross')
```

```python
# insert: 특정 위치로 열 삽입
profit_col_idx = movie_df.columns.get_loc('gross') + 1  # get_loc('column_name'): column_name의 열 위치 반환
movie_df.insert(loc=profit_col_idx,
                column='profit',
                value=movie_df['gross'] - movie_df['budget'])
```
