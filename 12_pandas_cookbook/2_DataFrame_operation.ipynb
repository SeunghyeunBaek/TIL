## 열 이름 정렬하기
* numeric, categorical 로 분류
* 그룹화
* 중요도 순으로 정렬, categorical > numeric

```python
# index 회복
movie_df.reset_index(inplace=True)
# 모든 열 출력
movie_df.columns
```

```python
# categorical, numeric columns

# categorical
cate_core = ['movie_title', 'title_year', 'content_rating', 'genres']
cate_people = ['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name']
cate_other = ['color', 'country', 'language', 'plot_keywords', 'movie_imdb_link']

# numeric
numeric_fb = ['director_facebook_likes', 'actor_1_facebook_likes',
             'actor_2_facebook_likes', 'actor_3_facebook_likes', 
             'cast_total_facebook_likes', 'movie_facebook_likes']
numeric_finance = ['budget', 'gross']
numeric_num_reviews = ['num_voted_users', 'num_user_for_reviews', 'num_critic_for_reviews']
numeric_other = ['imdb_score', 'duration', 'aspect_ratio', 'facenumber_in_poster']
```

```python
# 정렬 cate ~ num
new_col_order = cate_core + cate_people + cate_other + \
numeric_finance + numeric_num_reviews + numeric_other

# hard 오타검사
for new_col_name in new_col_order:
    cnt = 0
    for ex_col_name in movie_df.columns:
        if new_col_name == ex_col_name:
            cnt += 1
    if cnt == 0:
        print(f'=={new_col_name}, error')
    else:
        print(f'{new_col_name}, checked')
```

```python
movie_col_ordered_df = movie_df[new_col_order]
movie_col_ordered_df.head() 
```

## DataFrame Operator
* UGDS_: 학부생의 인종별 비율

```python
# load & check data
file_name = 'college.csv'
college_df = pd.read_csv(root+file_name, index_col='INSTNM')
```

```python
# UGDS_로 시작하는 열 선택
ugds_df = college_df.filter(regex='UGDS_\w*')
ugds_df.head()
```

```python
# 각 열 반올림
ugds_round_df = (ugds_df + .0001).round(2)
ugds_round_df.head()
```

## 누락값 비교

```python
# np.nan 은 비교연산자에 대해 False 반환
print(np.nan == np.nan)

# 모두 False기떄문에 누락값을 연산하지 않는다.
print((ugds_df == np.nan).sum())

# 누락값이 포함돼 False로 출력
print((ugds_df == ugds_df).all()) 
```

```python
# 누락값 계산은 isnull(), 비교는 equal 연산 사용
display(ugds_df.isnull().mean())
display(ugds_df.equals(ugds_df))
```

```python
# 결측치 열 삭제
ugds_df = ugds_df.dropna(how='all', axis=0)
ugds_df.isnull().sum()
```

```python
# 0.15보다 큰 값을 True로 변환하고 행별로 더한다.
diversity_metric = ugds_df.ge(.15).sum(axis=1).sort_values(ascending=False)
diversity_metric
```
