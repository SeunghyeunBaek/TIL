# RSS

국제 표준으로 양식이 정해진 문서

&lt;rss version=2.0&gt; 로 시작하면 RSS 문서다 

rss 문서의 구조와 태그는 정해져 있는 태그를 사용하고 있고 항목 하나의 데이터를 가지고 있는 item 태그 내부는 자유롭게 구성돼있다.

따라서 item 태그 내부만 살펴보면 쉽게 데이터를 가져올 수 있다.

```python
import requests
import bs4

import feedparser
```

```python
site = 'http://aladin.co.kr/rss/special_new/351'
res = requests.get(site)
soup = bs4.BeautifulSoup(res.text,'lxml-xml')
```

```python
# 전체 Title, link 추출

title = soup.select('channel > title')[0].text.strip()
link = soup.select('channel > link')[0].text.strip()

# item_list
item_list = soup.select('item')
for item_tag in item_list:
    item_title = item_tag.select('title')[0].text.strip()
    item_link = item_tag.select('link')[0].text.strip()
    print(f'{item_title} \n{item_link} \n--------------------------------------')
```
